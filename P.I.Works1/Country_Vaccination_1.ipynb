```python
new_df  = pd.DataFrame()
for ulke in list((df["country"].unique())):
    w = df.loc[df['country'] == ulke ]
    w = w.fillna(w.min()["daily_vaccinations"])
    new_df = pd.concat([new_df,w])
new_df
```
